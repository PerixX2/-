```python
import numpy as np
import matplotlib.pyplot as plt

x_min, x_max = -2, 2
y_min, y_max = -2, 2
step = 0.01

x, y = np.meshgrid(np.arange(x_min, x_max, step),
                   np.arange(y_min, y_max, step))

eq1 = np.sin(x + 1) - y - 1
eq2 = 2 * x + np.cos(y) - 2
fig, ax = plt.subplots(figsize=(8, 8))
ax.contour(x, y, eq1, levels=[0], colors='red')
ax.contour(x, y, eq2, levels=[0], colors='blue')

ax.set_xlim([x_min, x_max])
ax.set_ylim([y_min, y_max])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Графік системи рівнянь (Варіант 21)')
plt.grid(True)

plt.show()
```

```python
import numpy as np
from scipy import optimize
import math

x0 = 0.4
y0 = -0.1
epsilon = 0.001

def f1_iter(y):
    return 1 - 0.5 * math.cos(y)

def f2_iter(x):
    return math.sin(x + 1) - 1

def simple_iteration(x, y, e):
    xn = x
    yn = y
    xn1 = f1_iter(yn)
    yn1 = f2_iter(xn)
    n = 1

    while (abs(xn1 - xn) >= e) or (abs(yn1 - yn) >= e):
        xn = xn1
        yn = yn1
        xn1 = f1_iter(yn)
        yn1 = f2_iter(xn)
        n += 1

    print('Simple iteration:')
    print(f'x = {xn1}\ny = {yn1}\nThe amount of iterations = {n}')

simple_iteration(x0, y0, epsilon)

def f3_check(vars):
    x, y = vars
    return [math.sin(x + 1) - y - 1, 2 * x + math.cos(y) - 2]

s = optimize.root(f3_check, [x0, y0], method='hybr')
print('\nCheck', s.x)
```
