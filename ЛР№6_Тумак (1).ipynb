```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([-4, -2, 1, 3], dtype=float)
y = np.array([-8, 10, -8, 20], dtype=float)
x_test_points = np.array([-3, -1, 0.5, 2.5], dtype=float)
def lagrange_interpolation(x, y, x_test):
    n = len(x)
    p = np.zeros(n)
    for i in range(n):
        p_i = 1
        for j in range(n):
            if i != j:
                p_i *= (x_test - x[j])/(x[i] - x[j])
        p[i] = p_i
    return np.dot(y, p)
y_test_values = [lagrange_interpolation(x, y, xi) for xi in x_test_points]


print("Результати для тестових точок (Варіант №21)")
for xi, yi in zip(x_test_points, y_test_values):
    print(f"Значення функції у точці x = {xi: 4} = {yi.round(4)}")
print("-")
xnew = np.linspace(np.min(x), np.max(x), 100)
ynew = [lagrange_interpolation(x, y, i) for i in xnew]

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o', label='Базові вузли')
plt.plot(xnew, ynew, 'b-', label='Інтерполяційний поліном Лагранжа')
plt.plot(x_test_points, y_test_values, 'ro', markersize=8, label='Обчислені тестові точки')

for xi, yi in zip(x_test_points, y_test_values):
    plt.text(xi, yi + 1, f'{yi:.2f}', color='red', fontsize=9, ha='center')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Інтерполяція Лагранжа (Варіант №21)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
# ВАРІАНТ 1
x = np.array([4, 5, 6, 7], dtype=float)
def f_x(x):
    return np.sqrt(x) / (4 - np.sqrt(x) + np.log10(1 + x))
y_data = f_x(x)

print('x =', x)
print('f_x(x) =', y_data)
print('shapes:', x.shape, y_data.shape)

p = lagrange(x, y_data)
print('\nПоліном Лагранжа p(x):')
print(p)
print('Ступінь полінома:', p.order)

x_test = 4.3
y_test = p(x_test)
print(f'\nЗначення полінома в заданій точці x = {x_test}: {y_test:.5f}')

z = np.linspace(np.min(x), np.max(x), 100)

plt.plot(z, p(z), '-', label='$L(x)$')
plt.plot(x, y_data, 'o', label='data')
plt.plot(x_test, y_test, 'ro', label=f'x={x_test}')

plt.grid(True)
plt.xlabel('x'); plt.ylabel('y')
plt.legend()
plt.show()
```
