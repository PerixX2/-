```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.array([2.4, 2.6, 2.8, 3.0, 3.2, 3.4, 3.6, 3.8, 4.0, 4.2, 4.4, 4.6])
y = np.array([3.526, 3.782, 3.945, 4.043, 4.104, 4.155, 4.222, 4.331, 4.507, 4.775, 5.159, 5.683])

h = 0.2
n = len(x)

delta = np.zeros((n, n))
delta[:, 0] = y
for j in range(1, n):
    for i in range(n - j):
        delta[i, j] = delta[i + 1, j - 1] - delta[i, j - 1]

df_diff = pd.DataFrame(delta)
print("Таблиця скінченних різниць:")
print(df_diff.round(4).head(10))

idx = 10
dy_1 = delta[idx-1, 1]
d2y_2 = delta[idx-2, 2]
d3y_3 = delta[idx-3, 3]
d4y_4 = delta[idx-4, 4]
d5y_5 = delta[idx-5, 5]

y_prime_newton = (1/h) * (dy_1 + d2y_2/2 + d3y_3/3 + d4y_4/4 + d5y_5/5)
y_double_prime_newton = (1/h**2) * (d2y_2 + d3y_3 + (11/12)*d4y_4 + (5/6)*d5y_5)

first_diff = np.diff(y) / h
second_diff = np.diff(first_diff) / h

y_prime_def = first_diff[10]
y_double_prime_def = second_diff[9]

print(f" Значення у точці x = 4.4")
print(f"y'(4.4) за Ньютоном: {round(y_prime_newton, 4)}")
print(f"y''(4.4) за Ньютоном: {round(y_double_prime_newton, 4)}")
print(f"y'(4.4) за визначенням: {round(y_prime_def, 4)}")
print(f"y''(4.4) за визначенням: {round(y_double_prime_def, 4)}")
```
