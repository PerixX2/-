```python
import numpy as np

def simple_iteration_method(alpha, beta, epsilon=1e-3, max_iter=1000):
    x_curr = np.copy(beta)

    print("Початкове наближення x(0):", np.round(x_curr, 5))
    print("-" * 50)

    for k in range(1, max_iter + 1):
        x_next = np.dot(alpha, x_curr) + beta
        if k <= 3:
            print(f"Ітерація {k}: x({k}) = {np.round(x_next, 5)}")
        norm_diff = np.max(np.abs(x_next - x_curr))

        if norm_diff < epsilon:
            print("-" * 50)
            print(f"Досягнуто заданої точності epsilon = {epsilon} на {k}-й ітерації.")
            return x_next

        x_curr = x_next

    print("\nМетод не зійшовся за вказану кількість ітерацій.")
    return x_curr

def main():
    alpha = np.array([
        [0.13,  0.23, -0.44, -0.05],
        [0.24,  0.00, -0.31,  0.15],
        [0.06,  0.15,  0.00, -0.23],
        [0.72, -0.08, -0.05,  0.00]
    ])

    beta = np.array([2.13, -0.18, 1.44, 2.42])
    x_final = simple_iteration_method(alpha, beta, epsilon=1e-8)

    print("Фінальний розв'язок (корені системи):", np.round(x_final, 5))

if __name__ == "__main__":
    main()
```

```python
import numpy as np
alpha = np.array([
    [0.13,  0.23, -0.44, -0.05],
    [0.24,  0.00, -0.31,  0.15],
    [0.06,  0.15,  0.00, -0.23],
    [0.72, -0.08, -0.05,  0.00]
])

beta = np.array([2.13, -0.18, 1.44, 2.42])

E = np.eye(len(alpha))
A = E - alpha

x_exact = np.linalg.solve(A, beta)

print("Точний розв'язок для перевірки:")
print(np.round(x_exact, 5))
```
