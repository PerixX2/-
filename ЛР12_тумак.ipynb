Завдання 1. Метод прямокутників


```python
from scipy import integrate
import numpy as np

def left_rec(f, a, b, n):
    h = (b - a) / n
    sum = 0
    for i in range(0, n):
        sum += f(a + i * h)
    return sum * h

def right_rec(f, a, b, n):
    h = (b - a) / n
    sum = 0
    for i in range(1, n + 1):
        sum += f(a + i * h)
    return sum * h

def aver_rec(f, a, b, n):
    h = (b - a) / n
    sum = 0
    for i in range(0, n):
        sum += f(a + (i + 0.5) * h)
    return sum * h

n = 10
print("Інтеграл 1")
a1, b1 = 0.6, 1.4

def f1(x):
    return 1 / np.sqrt(12 * x + 0.5)

print("Left rectangle:   ", round(left_rec(f1, a1, b1, n), 4))
print("Right rectangle:  ", round(right_rec(f1, a1, b1, n), 4))
print("Average rectangle:", round(aver_rec(f1, a1, b1, n), 4))

v1, err1 = integrate.quad(f1, a1, b1)
print("SciPy quad check: ", round(v1, 4))
print()

print("Інтеграл 2")
a2, b2 = 1.3, 2.1

def f2(x):
    return np.sin(x**2 - 1) / (2 * np.sqrt(x))

print("Left rectangle:   ", round(left_rec(f2, a2, b2, n), 4))
print("Right rectangle:  ", round(right_rec(f2, a2, b2, n), 4))
print("Average rectangle:", round(aver_rec(f2, a2, b2, n), 4))

v2, err2 = integrate.quad(f2, a2, b2)
print("SciPy quad check: ", round(v2, 4))
print()

print("Інтеграл 3")
a3, b3 = 0.6, 1.5

def f3(x):
    return 1 / np.sqrt(1 + 2 * x**2)

print("Left rectangle:   ", round(left_rec(f3, a3, b3, n), 4))
print("Right rectangle:  ", round(right_rec(f3, a3, b3, n), 4))
print("Average rectangle:", round(aver_rec(f3, a3, b3, n), 4))

v3, err3 = integrate.quad(f3, a3, b3)
print("SciPy quad check: ", round(v3, 4))
```

Завдання 2 Метод Сімпсона

```python
from scipy import integrate
import numpy as np

def simpson_rule(f, a, b, n):
    if n % 2 != 0:
        n += 1

    h = (b - a) / n
    integr = f(a) + f(b)

    for i in range(1, n):
        k = a + i * h
        if i % 2 == 0:
            integr += 2 * f(k)
        else:
            integr += 4 * f(k)

    integr *= h / 3
    return integr

n = 8

print("Інтеграл 1")
a1, b1 = 0.6, 1.4

def f1(x):
    return 1 / np.sqrt(12 * x + 0.5)

print("Simpson method:", round(simpson_rule(f1, a1, b1, n), 4))
v1, err1 = integrate.quad(f1, a1, b1)
print("SciPy quad check:", round(v1, 4))
print()

print("Інтеграл 2")
a2, b2 = 1.3, 2.1

def f2(x):
    return np.sin(x**2 - 1) / (2 * np.sqrt(x))

print("Simpson method:", round(simpson_rule(f2, a2, b2, n), 4))
v2, err2 = integrate.quad(f2, a2, b2)
print("SciPy quad check:", round(v2, 4))
print()

print("Інтеграл 3")
a3, b3 = 0.6, 1.5

def f3(x):
    return 1 / np.sqrt(1 + 2 * x**2)

print("Simpson method:", round(simpson_rule(f3, a3, b3, n), 4))
v3, err3 = integrate.quad(f3, a3, b3)
print("SciPy quad check:", round(v3, 4))
```

Завдання 3 Визначити інтеграм методом трапецій

```python
from scipy import integrate
import numpy as np

def trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    sum = 0.5 * (f(a) + f(b))
    x = a

    for i in range(1, n):
        x += h
        sum += f(x)

    return h * sum

n = 10

print("Інтеграл 1")
a1, b1 = 0.6, 1.4

def f1(x):
    return 1 / np.sqrt(12 * x + 0.5)

print("Trapezoidal method:", round(trapezoidal_rule(f1, a1, b1, n), 4))
v1, err1 = integrate.quad(f1, a1, b1)
print("SciPy quad check:  ", round(v1, 4))
print()

print("Інтеграл 2")
a2, b2 = 1.3, 2.1

def f2(x):
    return np.sin(x**2 - 1) / (2 * np.sqrt(x))

print("Trapezoidal method:", round(trapezoidal_rule(f2, a2, b2, n), 4))
v2, err2 = integrate.quad(f2, a2, b2)
print("SciPy quad check:  ", round(v2, 4))
print()

print("Інтеграл 3")
a3, b3 = 0.6, 1.5

def f3(x):
    return 1 / np.sqrt(1 + 2 * x**2)

print("Trapezoidal method:", round(trapezoidal_rule(f3, a3, b3, n), 4))
v3, err3 = integrate.quad(f3, a3, b3)
print("SciPy quad check:  ", round(v3, 4))
```
